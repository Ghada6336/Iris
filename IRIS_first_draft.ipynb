# Exploratory Data Analysis (EDA) on Iris Dataset

Import necessary libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the Iris dataset

```python
iris = sns.load_dataset('iris')
```

Display the first few rows of the dataset

```python
print("First few rows of the Iris dataset:")
print(iris.head())
```

Summary statistics of the dataset

```python
print("\nSummary statistics of the Iris dataset:")
print(iris.describe())
```

Pairplot to visualize relationships between features

```python
sns.pairplot(iris, hue='species', diag_kind='kde')
plt.suptitle("Pairplot of Iris Dataset")
plt.show()
```

Boxplot to visualize distributions of features by species

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title("Distribution of Sepal Length by Species")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x='species', y='sepal_width')
plt.title("Distribution of Sepal Width by Species")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x='species', y='petal_length')
plt.title("Distribution of Petal Length by Species")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x='species', y='petal_width')
plt.title("Distribution of Petal Width by Species")
plt.show()
```
